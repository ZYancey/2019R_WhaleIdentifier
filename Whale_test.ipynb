**Whale Song Audio File converted to Features Test**

This is a quick test to validate my idea that wav audio files of whale songs can be transformed into numerical features that can be used an input into various machine learning algorithms.

```python
# Here are the necessary imports.
import pandas as pd
import os
import librosa
import librosa.display
import numpy as np
import IPython.display as ipd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Here is one of the wav files of a humpback whale song
ipd.Audio('humpback.wav')
```

```python
signal, sample_rate = librosa.load('humpback.wav',sr = None)
# sr = None makes sure the native sample rate of audio is used.
```

The signal or audio_data are integers with only one dimension.

```python
print(signal)
```

```python
print(sample_rate)
```

```python
# Short-Time Fourier Transform
print(librosa.stft(signal))
```

```python
# A spectrogram displays the strength of the signal or loudness(in decibels)
spectrogram  = librosa.amplitude_to_db(np.abs(librosa.stft(signal)), ref=np.max)
print(spectrogram )
```

```python
file_name='humpback.wav'
audio_data, sampling_rate = librosa.load(file_name)
```

```python
# Using the Mel Frequency Cepstral Coefficient (MFCC) algorithm,
# we can break our sound file into different numerical features
mfccs = librosa.feature.mfcc(y=audio_data, sr=sampling_rate, n_mfcc=10)
print(mfccs)
print(len(mfccs))
```

```python
# We can take the mean of those features
mfccs_scaled_features = np.mean(mfccs.T,axis=0)
print(mfccs_scaled_features)
```
